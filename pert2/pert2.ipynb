```python
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)

plt.pie(
    votes,
    labels=flavors,
)
plt.show()
```

```python
import matplotlib.pyplot as plt

matkul = ('Basis Data', 'ML', 'Pemrograman WEB', 'Jaringan Komputer')
nilai = (90, 85, 75, 80)
colors =  ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0, 0, 0.1, 0)


plt.pie(
    nilai,
    labels=matkul,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
)
plt.show()
```
